```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.style.use('seaborn-v0_8-deep')
plt.plot(x,y)

```

```python
z=np.cos(x)
plt.style.use('fivethirtyeight')
plt.plot(x,z)
```

```python
p=np.exp(x)
plt.style.use('ggplot')
plt.plot(x,p)
```
