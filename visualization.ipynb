```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.style.use('seaborn-v0_8-deep')
plt.plot(x,y)

```

```python
z=np.cos(x)
plt.style.use('fivethirtyeight')
plt.plot(x,z)
```

```python
p=np.exp(x)
plt.style.use('ggplot')
plt.plot(x,p)
```

```python
fig=plt.figure(figsize=(6,4))
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,marker='H')
```

```python
fig=plt.figure(figsize=(6,4))
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,marker='o',linestyle=':',markersize=4,markerfacecolor='blue',markeredgecolor='green',markeredgewidth=1,linewidth='2')
```

```python
fig=plt.figure(figsize=(6,4))
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,linestyle='--',color='blue',linewidth='1',label='sin(x)')
plt.title('time series')
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```

```python
fig=plt.figure(figsize=(6,4))
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,linestyle='--',color='blue',linewidth='1',label='sin(x)')
plt.plot(x,z,linestyle='--',color='red',linewidth='1',label='cos(x)')
plt.title('time series',loc='left',pad='10')
plt.xlabel('time')
plt.ylabel('value')
plt.xlim(2,6) #to zoom in or zoom out at particular section or graph
plt.ylim(-1,-1)
plt.legend()
```

```python
rainfall=[.8,0.0,.6,0.3,.1,0.0,0.0,1.2,1.5,0.0,0.3,0.5]
months=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
fig=plt.figure(figsize=(6,4))
plt.bar(months,rainfall,color='blue',hatch='//')
plt.title('rainfall in 2019')
plt.xlabel('month')
plt.ylabel('rainfall(mm)')
```
