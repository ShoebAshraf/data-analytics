```python
def clean_price(price):
    if price is None:
        return np.nan
    elif isinstance(price,str):
        price=price.replace(',', '')
        if price:
            price=float(price)
        else:
            price=np.nan
        return price
    else:
        return price
```

```python
df['actual_price']=df['actual_price'].apply(clean_price)
```

```python
df['actual_price'].plot.hist(bins=50)
```

```python
def clean_rating(value):
    if value is None:
        return np.nan
    elif isinstance(value,str):
        if value:
            value= float(value)
        else:
            value=np.nan
        return value
    else:
        return value

    
```

```python
df['average_rating']=df['average_rating'].apply(clean_rating)
```

```python
df['average_rating'].plot()
```

```python
def clean_discount(value):
    if value is None:
        return np.nan
    elif isinstance(value,str):
        value=value.replace('% off', '')
        if value:
            value=float(value)
        else:
            value=np.nan
        return value
    else:
        return value
```

```python
df['discount']=df['discount'].apply(clean_discount)
```

```python
def clean_sp(value):
    if value is None:
        return np.nan
    elif isinstance(value,str):
        value=value.replace(',', '')
        if value:
            value=float(value)
        else:
            value=np.nan
        return value
    else:
        return value
```

```python
df['selling_price']=df['selling_price'].apply(clean_sp)
```
